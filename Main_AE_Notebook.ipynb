```python
import librosa
import IPython.display as ipd
from Hyperparameters import sample_rate, n_fft, hop_size, n_fft, top_db
from Util import map_to_range

def gen(path):
    result = main.generate(path)
    inv_db_final = map_to_range(result, 0, 1, -top_db, 0)
    inv_pow = librosa.db_to_power(inv_db_final)
    sig_result = librosa.feature.inverse.mel_to_audio(inv_pow, sample_rate, n_fft, hop_size, n_fft)
    sig_result = librosa.util.normalize(sig_result)
    return sig_result

#ipd.Audio(gen("data/piano/chpn_op7_1.wav"), rate=sample_rate)
#ipd.Audio(gen("data/piano/chp_op18.wav"), rate=sample_rate)
#ipd.Audio(gen("data/piano/chpn_op53.wav"), rate=sample_rate)
ipd.Audio(gen("data/piano/chpn-p13.wav"), rate=sample_rate)
```
