```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300 # Set high dpi for high-quality plots

# Plot autoencoder losses
ae_train_losses = np.load("train_losses_ae.npy")
ae_test_losses = np.load("test_losses_ae.npy")

plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.plot(np.arange(ae_train_losses.size), ae_train_losses)
plt.title("Autoencoder training losses over epochs")

plt.subplot(2, 1, 2)
plt.plot(np.arange(ae_test_losses.size), ae_test_losses)
plt.title("Autoencoder test losses over epochs")
plt.show()

# Plot U-Net losses
unet_train_losses = np.load("train_losses_unet.npy")
unet_test_losses = np.load("test_losses_unet.npy")

plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
# Use log scale for y-axis
plt.yscale('log')
plt.plot(np.arange(unet_train_losses.size), unet_train_losses)
plt.title("U-Net training losses over epochs")

plt.subplot(2, 1, 2)
plt.yscale('log')
plt.plot(np.arange(unet_test_losses.size), unet_test_losses)
plt.title("U-Net test losses over epochs")
plt.show()
```
